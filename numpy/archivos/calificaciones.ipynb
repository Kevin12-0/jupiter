```python
import numpy as np
import os
file = os.path.join('calificaciones_v1.csv')
```

```python
array = np.loadtxt(file, dtype=str, delimiter=',')
array
```

```python
filter = np.array([False,True])
new = array[:,filter]
```

```python
new = new.reshape(48,9)
new
```

```python
array_2 = np.array([['Matricula','Nombre','Grupo','M1','M2','M3','M4','M5','estado','promedio'],new],dtype = object)
array_2
```

```python
with open('calificaciones_finales.csv', 'a') as f:
    f.write(str(array_2[0]))
```

```python
counter = 3
plus = 0
row = 0
while row <= 10:
    with open('calificaciones_finales.csv', 'a') as f:
        while counter <=7:
            array_3 = new[row]
            plus = plus + int(array_3.item(counter))
            counter += 1
            prom = plus/5
        f.write('\n'+str(array_3)+','+str(prom))
        row += 1
        counter = 3
        plus = 0
```

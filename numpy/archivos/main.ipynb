```python
import numpy as np
import os
```

```python
file = os.path.join('calificaciones_v1.csv')
```

```python
array = np.loadtxt(file, dtype = str, delimiter = ',')
array
```

```python
array = array.reshape(48,18)
array
```

```python
array[0:1]
```

```python
count = 7
column = 0
row = 1
suma  = 0 
while column <= 15:
    new = array[column:row]
    print(new)
    column +=1
    row +=1
    while count <= 15:
        suma = suma + int(new.item(count))
        count +=2
    promedio = suma / 5
    print('Promedio: '+ str(promedio))
    count = 7
    suma = 0
```
